## Gradient descent

```python
import numpy as np
from sympy import diff , log , exp
from sympy.abc import x
```

```python
expr = exp(x) - log(x)
dy_dx = diff(expr,x)

xi = 1.5
eta = 0.29
for i in range(1):
    new_x = xi - eta*dy_dx.subs([(x,xi)])
    print("xi = ",new_x)
    print("f(xi) = ",expr.subs([(x,new_x)]))
    xi = new_x
```
