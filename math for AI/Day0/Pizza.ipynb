```python
import pandas as pd
import numpy as np
```

```python
pizza = pd.read_csv('pizza.csv')
xi = pizza['Promote']
yi = pizza['Sales']
```

```python
x = pizza['Promote']
y = pizza['Sales']
xi = ((x-np.min(x))/(np.max(x)-np.min(x)))

yi = ((y-np.min(y))/(np.max(y)-np.min(y)))
```

```python
eta = 0.04
b0 = 0.4
w0 = -0.6

ycap = xi*w0 + b0

for i in range(1,501):

    db = -(1/len(xi))*np.sum((yi-ycap))     
    dw = -(1/len(xi))*np.sum((yi-ycap)*xi)  
        
    new_b = b0 - eta*db
    new_w = w0 - eta*dw
                    
    b0 = new_b
    w0 = new_w
    ycap = xi*w0 + b0
    
    print('L = ', (1/(2*len(xi))*np.sum((yi-ycap)**2)))
    print('w = ',w0,'; b =', b0)
```

### Gradient descent for two independent variables and one dependent variable

```python
db = pd.read_csv('Insure_auto.csv')
db.head()
```

```python
x2 = db['Automobile'].astype(int)
y = db['Operating_Cost'].astype(int)
x1 = db['Home'].astype(int)
x1 = ((x1-np.min(x1))/(np.max(x1)-np.min(x1)))
x2 = ((x2-np.min(x2))/(np.max(x2)-np.min(x2)))
yi = ((y-np.min(y))/(np.max(y)-np.min(y)))
```

```python
eta = 0.4
b0 = .4
w1 = -0.6
w2 = 0.5
nobs = db.shape[0]
ycap = x1*w1 + x2*w2 + b0

for i in range(1,1001):

    db = -(1/nobs)*np.sum((yi-ycap))     
    dw1 = -(1/nobs)*np.sum((yi-ycap)*x1)  
    dw2 = -(1/nobs)*np.sum((yi-ycap)*x2)  
        
    new_b = b0 - eta*db
    new_w1 = w1 - eta*dw1
    new_w2 = w2 - eta*dw2

    b0 = new_b
    w1 = new_w1
    w2 = new_w2
    ycap = x1*w1 + x2*w2 + b0
    
    print('L = ', (1/(2*nobs)*np.sum((yi-ycap)**2)))

    # print('w = ',w0,'; b =', b0)
    print(f" w1 = {w1} ; w2 = {w2} ; b0 = {b0}")
```
