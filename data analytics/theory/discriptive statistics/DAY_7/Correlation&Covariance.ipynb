### Example 2

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head(10)
```

```python
correlation_coefficient = df['reading score'].corr(df['writing score'])

print(f'Correlation coefficient : {correlation_coefficient} ' )
```

```python
plt.scatter(df['reading score'],df['writing score'])
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title(f'Scatter Plot (Correlation: {correlation_coefficient: .2f})')
# Identify potential outliers(eg., values with residual greater than 2 times the standard deviation)
residuals = df['writing score'] - df['reading score']
print(residuals)
std_deviation = residuals.std()
outliers = df[abs(residuals) > 2 * std_deviation]
# Highlight potential outliers on the scatter plot
plt.scatter(outliers['reading score'],outliers['writing score'],color='red',label='Outliers')
plt.legend()
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('loan.csv')
df.head(10)
```

```python
plt.scatter(df['age'],df['income'])
plt.xlabel('age')
plt.ylabel('income')
plt.title(f'Scatter Plot (Correlation: {correlation_coefficient: .2f})')
# Identify potential outliers(eg., values with residual greater than 2 times the standard deviation)
residuals = df['income'] - df['age']
print(residuals)
std_deviation = residuals.std()
outliers = df[abs(residuals) > 2 * std_deviation]
# Highlight potential outliers on the scatter plot
plt.scatter(outliers['age'],outliers['income'],color='red',label='Outliers')
plt.legend()
plt.show()
```

## Covariance 

### Example 1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
advert = pd.read_csv('Advertising.csv')
advert.sample(10)
```

```python
# Covariance Coefficient
print(advert['TV'].cov(advert['Newspaper']))
# OR
print(np.cov(advert['TV'].cov(advert['Newspaper'])))
```

```python
print(advert.cov())
```

```python
dataplot = sns.heatmap(advert.cov(),annot=True,fmt='g')
plt.show()
# the amount spent on x and y combination 
```

```python
print(advert['TV'].min())
print(advert['Newspaper'].min())
print(advert['Radio'].min())
print(advert['Sales'].min())
```

```python
import yfinance as yf
```

```python
tickers = ['TCS.NS','ICICIBANK.NS','RELIANCE.NS','BHARTIARTL.NS','ITC.NS','MARUTI.NS','BAJFINANCE.NS']
```

```python
df = yf.download(tickers,period='3y')
```

```python
df
```

```python
close = df['Close'] # get the closing price
close
```

the code calculate the daily percentage returns for each stock in close dataframe . the returns are normalized by dividing each closing price by the initial closing price and multiplying it by 100

```python
norm = close.div(close.iloc[0]).mul(100) 
# .div() --> scales all val (summation(xi/x*100) i for 1 to n) x will be inital val of column
norm
```

```python
ret = close.pct_change().dropna()
```

```python
print(ret)
```

```python
print(ret.cov())
print(ret.cov())
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(ret.corr(),cmap='Blues',annot=True)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(ret.cov(),cmap='Blues',annot=True)
plt.show()
```
