# __Skewness__

#### Example 1

* Imports

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

Genrate a normal distribution with 10000 data points plot it and show statistics

```python
np.random.seed(33)
norm_sample = np.random.randn(10000)
sns.histplot(x = norm_sample, kde=True)
plt.show()
```

```python
print('Mean = ',np.mean(norm_sample))
print('Median = ',np.median(norm_sample))
print('standerd deviation = ',np.std(norm_sample))
```

* genrate a sample of 10000 data points from right-skewed distribution
* first parameter if 0 means no skewness, +ve means possitive skewness , -ve means nigative skewness
* this mean the genrated datawill have a positive skew i.e long tail will be on the right side and mejority of the data points will be on the left side
* 10 here is a just a 10 units, if you increase it the right skewness will increse even further

```python
skewed_data_R = stats.skewnorm.rvs(10,size = 10000)
sns.histplot(x = skewed_data_R, kde=True)
```

```python
print('Mean = ',np.mean(skewed_data_R))
print('Median = ',np.median(skewed_data_R))
print('standerd deviation = ',np.std(skewed_data_R))
```

* genrate a sample of 10000 data points from left-skewed distribution
* first parameter if 0 means no skewness, +ve means possitive skewness , -ve means nigative skewness
* this mean the genrated datawill have a Nigative skew i.e long tail will be on the left side and mejority of the data points will be on the right side
* -10 here is a just a 10 units, if you decrease it the left skewness will increse even further

```python
skewed_data_L = stats.skewnorm.rvs(-10,size = 10000)
sns.histplot(x = skewed_data_L, kde=True)
```

```python
print('Mean = ',np.mean(skewed_data_L))
print('Median = ',np.median(skewed_data_L))
print('standerd deviation = ',np.std(skewed_data_L))
```

__Q-Q Plot__
* Q-Q plot of normally distributed data

```python

stats.probplot(norm_sample,plot=plt)
plt.show()
```

* Q-Q Plot of Right skwewed data

```python

stats.probplot(skewed_data_R,plot=plt)
plt.show()
```

* Q-Q Plot of Left skwewed data

```python
stats.probplot(skewed_data_L,plot=plt)
plt.show()
```

Print the Skewness of the each data set

```python
print(stats.skew(norm_sample))
print(stats.skew(skewed_data_R))
print(stats.skew(skewed_data_L))
```

* perform tha skewness test on each dataset
* return a numerical value reprenting the skewness of the dataset

```python
print(stats.skewtest(norm_sample))
print(stats.skewtest(skewed_data_R))
print(stats.skewtest(skewed_data_L))
```

__output : Statistic__: The Statistic value is the test of statistic calculated by the skewness test. It quntifies the degree of skewness in the dataset.
if negative, it indicates that the dataset is left-skewed or nigatively skewed.

the p-value measure the evidence aginst the null hypothesis. :-
* A small p-value (tipically<0.05) suggests that there is a strong evidence to reject null hypothesis, indicating that the data is significantly skewed.
* A large p-value (tipically>0.05) suggests that there is a not enough evidence to reject null hypothesis, indicating that the data may not be significantly skewed.
