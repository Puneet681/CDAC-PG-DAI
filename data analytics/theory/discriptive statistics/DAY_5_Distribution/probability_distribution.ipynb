```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import empiricaldist
from statadict import parse_stata_dict
```

```python
dict_file = 'GSS.dct'
data_file = 'GSS.dat.gz'
```

```python
stata_dict = parse_stata_dict(dict_file)
import gzip

fp = gzip.open(data_file)


# Convert the file into Pandas DataFrame
# Fwf ->Fixed Width File  FP-> File Pointer
gss = pd.read_fwf(fp, names = stata_dict.names, colspecs = stata_dict.colspecs)
gss
```

```python
print(gss['EDUC'].value_counts().sort_index())
```

```python
import numpy as np
educ = gss['EDUC'].replace([98,99],np.nan)
```

```python
import matplotlib.pyplot as plt
```

```python
educ.hist(grid = False)
plt.xlabel('Years of education')
plt.ylabel('Number of respondents')
plt.title('Histogram of Education Level')
plt.show()
```

```python
from empiricaldist import Pmf
pmf_educ = Pmf.from_seq(educ,normalize=False)
print(type(pmf_educ))
print(pmf_educ.head())
print(pmf_educ.tail())
print(pmf_educ[20])
```

```python
pmf_educ_norm = Pmf.from_seq(educ,normalize=True)
print(pmf_educ_norm.head())
print(pmf_educ_norm[12]) 
```

```python
pmf_educ_norm.bar(label='EDUC')
plt.xlabel('Years of education')
plt.xticks(range(0,21,4))
plt.ylabel('PMF')
plt.title('Distribution of Years of Education')
plt.legend();
plt.show()
```

```python
# IQR inter quirtile Range
print(q3-q1)
```

```python

# create series for male and female respondents
male = (gss['SEX']==1)
female = (gss['SEX']==2)
```

```python
age = gss['AGE'].replace([99,98],np.nan)
age
```

```python
# Select Ages
male_a = age[male]
female_a = age[female]
male_a
```

```python
# plot PMF for Each
pmf_male_age = Pmf.from_seq(male_a)
pmf_male_age.plot()
pmf_female_age = Pmf.from_seq(female_a)
pmf_female_age.plot()
plt.xlabel('Age (years)')
plt.ylabel('PMF')

plt.title('Distribution of age by sex')
plt.legend()
plt.show()
```

```python
cdf_male_age = Cdf.from_seq(male_a)
cdf_male_age.plot()
cdf_female_age = Cdf.from_seq(female_a)
cdf_female_age.plot()
plt.xlabel('Age (years)')
plt.ylabel('CDF')

plt.title('Distribution of age by sex')
plt.legend()
plt.show()
```

```python
print(cdf_male_age(60),cdf_female_age(60))
```

```python
# Compaing male and female at 50th precentile 
print(cdf_male_age.inverse(0.5),cdf_female_age.inverse(0.5))
```

```python
print(1-cdf_male_age(80),1-cdf_female_age(80))
```

```python
pre95 = (gss['YEAR'] < 1995)
post95 = (gss['YEAR'] >= 1995)

income = gss['REALINC'].replace(0,np.nan)

Pmf.from_seq(income[pre95]).plot(label='Before 1995')
Pmf.from_seq(income[post95]).plot(label='After 1995')

plt.xlabel('Income (1986 USD)')
plt.ylabel('PMF')
plt.title("Distribution of Income (PMF)")
plt.legend();
plt.show()
```

```python
df = pd.merge(Pmf.from_seq(income[pre95]),Cdf.from_seq(income[pre95]),on='REALINC')
df.columns = ['PMF','CDF']
df
```

```python
Cdf.from_seq(income[pre95]).plot(label='Before 1995')
Cdf.from_seq(income[post95]).plot(label='After 1995')

plt.xlabel('Income (1986 USD)')
plt.ylabel('PMF')
plt.title("Distribution of Income (CDF)")
plt.legend();
plt.show()
```

```python
#Kernal Density Estimation or Probability Density Function (PDF)
age_data = gss['AGE']
age_data.head()
```

```python
plt.hist(age_data,bins=20,density=True,alpha=0.6,color='b',label='PDF')
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.title('PDF of Age in GSS Dataset')
plt.legend()
plt.show()
```

```python
pmf = age_data.value_counts(normalize=True).sort_index()
pmf
```

```python
plt.figure(figsize=(10,6))
plt.bar(pmf.index,pmf.values)
plt.xlabel('Age')
plt.ylabel('Probability')
plt.title('PMF of Age in GSS Dataset')
#plt.xticks(rotation=90)
plt.show()
```
