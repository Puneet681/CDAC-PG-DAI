#### __Central limit theorem__

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create a list to store the sample means
means = []
# reapete the experiment  100 times
for i in range(150):
    # draw a random sample of 5 values, with =values between 0 and 1
    arr = np.random.rand(5)
    # calculate the mean of samle
    s_mean = np.mean(arr)
    #add s_mean to the list
    means.append(s_mean)
    # plot it alowng with the mean of the distribution
sns.histplot(means,kde = True , #kerenel density estimate smoothens the plot
             bins=100 , color='darkblue')
# calculate the means
mean = sum(means)/len(means)
# Plot the mean over the distribution to get the sensce of central tencendy
plt.axvline(mean , color = 'k',linestyle = 'dashed',linewidth = 1)
min_ylim , max_ylim = plt.ylim()
plt.text(mean*1.1,max_ylim*0.9,'mean: {:.2f}'.format(mean))
plt.title('sampling distribution of sample mean of randomly genrated samples')
plt.xlabel('sample mean')
plt.ylabel('frequency')
plt.show()
```
