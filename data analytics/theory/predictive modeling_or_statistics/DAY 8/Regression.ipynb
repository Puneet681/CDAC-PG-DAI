### __linear Regression__

import pandas

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('salary.csv')
dataset
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size= 1/3 , random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_train,regressor.predict(x_train),color = 'blue')
plt.title('salary vs exprince(training set)')
plt.xlabel('year of exprince')
plt.ylabel('salary')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_test,y_test,color = 'green')
plt.plot(x_train,regressor.predict(x_train),color = 'blue')
plt.title('salary vs exprince(training set)')
plt.xlabel('year of exprince')
plt.ylabel('salary')
plt.show()
```

```python
new_salary_pred = regressor.predict([[15]])
print(f'the predicted salary of a person with 15 years of exprince is {new_salary_pred}')
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color = 'red')
plt.scatter(x_test,y_test,color = 'green')
plt.plot(x_train,regressor.predict(x_train),color = 'blue')
plt.title('salary vs exprince(training set)')
plt.xlabel('year of exprince')
plt.ylabel('salary')
plt.legend()
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
x = df.iloc[:,:3]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
df.head()
```

```python
plt.scatter(x_train['TV'],y_train,color='red')
plt.scatter(x_train['Radio'],y_train,color='Green')
plt.scatter(x_train['Newspaper'],y_train,color='Purple')
plt.scatter(x_test['TV'],y_test,color='red')
plt.scatter(x_test['Radio'],y_test,color='Green')
plt.scatter(x_test['Newspaper'],y_test,color='purple')
plt.plot(x_train,model.predict(x_train),color = 'blue')
plt.title('Advertisment vs Sales')
plt.xlabel('')
plt.ylabel('')
plt.legend(labels = ['TV','RADIO','Newspaper'])
plt.show()
```

* R Squared: R Square is the coefficient of determination. It tells us how many points fall on the regression line. The value of R Square is 90.11, which indicates that 90.11% of the data fit the regression model.

* Mean Absolute Error: Mean Absolute Error is the absolute difference between the actual or true values and the predicted values. The the value, the better is the model's performance. A mean absolute error can actully perfect predictor of the outputs. The mean absolut particular model is 1.227, which is pretty good as that your model is a ined for this to 0.

* Mean Square Error: Mean Square Error is calculated by Lng the average of the square of the difference between the original and predicted values of the data. The lower the value, the better is the model's performance. The mean square error obtained for this particular model is 2.636, which is pretty good.

* Root Mean Square Error: Root Mean Square Error is the standard deviation of the errors which occur when a prediction is made on a dataset. This is the same as Mean Squared Error, but the root of the value is considered while determining the accuracy of the model. The lower the value, the better is the model's performance. The root mean square error obtained for this particular model is

```python
from sklearn import metrics
meanAbErr =  metrics.mean_absolute_error(y_test,y_pred)
meanSqErr =  metrics.mean_squared_error(y_test,y_pred)
root_mean_error = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
print('R squared = {:.2f}'.format(model.score(x,y)))
print(f'Mean Absulute Error = {meanAbErr}')
print(f'Mean Squared Error = {meanSqErr}')
print(f'Root Mean Squared Error = {root_mean_error}')
```
