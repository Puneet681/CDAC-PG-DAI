### Example 1

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def roll_dice():
    # simulation of a rolling a dice twice, minimum number we get is 1, and maximum is6
    # add the results of the simulations, so possible values will be :
    # (1,1) or(1,2) ... (6,6).... sum will be between 2 and 12
    # run multiple times to verify
    s = (np.random.randint(1,7,2))
    # print(s)
    return np.sum(s)
print(roll_dice())
```

```python
# someone approchaes us and says that "i will give you 5 dollars if you get 7 and will take 1 dollar if other than 7 is got"
# how do we know waht will happen
# we crate our own "monte carle simulateion" like fuction
def monte_carlo_sim(runs=10):
    results = np.zeros(2)
    for i in range(runs):
        if roll_dice()==7:
            results[0] += 1
        else:
            results[1] += 1
    return results

print(monte_carlo_sim())
print(monte_carlo_sim())
```

```python
results = np.zeros(10)
lost = np.zeros(10)
for i in range(10):
    results[i] = monte_carlo_sim()[0]

print(results)
```

```python
results = np.zeros(1000)
lost = np.zeros(1000)
for i in range(1000):
    results[i],lost[i] = monte_carlo_sim()

print(results)
```

```python
plt.hist(lost,color='red')
plt.hist(results)
```

```python
print(results.mean())        #genral mean
print(results.mean()*5)      # what we will win on an avrage
print(1000-results.mean())   # what we will pay on avrage
print(results.mean()/1000)   # probability of the 'we will pay' result
# the last probability sghould be close to the theorotically probaillity of getting a 7 when we throw two dice
```
