a. Read the entire dataset using pandas dataframe and build the file by appending

```python
df1 = pd.read_csv('./tandem_data/201803-fordgobike-tripdata.csv')
df1.head(1)
```

```python
df2 = pd.read_csv('./tandem_data/201807-fordgobike-tripdata.csv')
df2.head(1)
```

```python
df3 = pd.read_csv('./tandem_data/201811-fordgobike-tripdata.csv')
df3.head(1)
```

```python
newdf = pd.concat([df1,df2,df3,df4])
newdf.sample(5)
```

```python
newdf.info()
```

```python
newdf.isna().sum()
```

b. What was the time duration for the average trip taken? (Use plotting techniques and
write your observations for the same.)

```python
newdf['locations'] = (newdf['start_station_name']) +" to "+(newdf['end_station_name'] )
newplot = pd.DataFrame(round(newdf.groupby('locations')['duration_sec'].mean(),2))
newplot
```

```python
high100 = newplot.sort_values(ascending=False,by='duration_sec').head(100)
high100.head(1)
```

```python
low100 = newplot.sort_values(ascending=False,by='duration_sec').tail(100)
high100 = newplot.sort_values(ascending=False,by='duration_sec').head(100)
```

```python
plt.figure(figsize=(22,5))
sns.barplot(x=high100.index, y=high100['duration_sec'])
plt.xticks(rotation=90);
plt.title("Duration For Lowest 100 Trips")
```

```python
plt.figure(figsize=(22,5))
sns.barplot(x=low100.index, y=low100['duration_sec'])
plt.xticks(rotation=90);
plt.title("Duration For Lowest 100 Trips")
```

c. Is there any correlation between the months and the trip duration? Use scatter plot
and comment your observations from the plot

```python
newdf['start_time'] = pd.to_datetime(newdf['start_time'])
newdf['months'] = newdf['start_time'].dt.month
```

```python
heatdf = pd.DataFrame(newdf.groupby('months')['duration_sec'].mean())
heatdf
```

```python
plt.scatter(x=heatdf.index,y=heatdf['duration_sec'])
```

```python
## Conclusion - 
## During Month 1 to 7 - Longer Trip Duration
## During Month 1 to 12 - Shorter Trip Duration
```

d. Do seasons affect the duration of a trip ? Comment your observations.


```python
heatdf = pd.DataFrame(newdf.groupby('months')['duration_sec'].mean())
heatdf
```

```python
plt.scatter(x=heatdf.index,y=heatdf['duration_sec'])
```

```python
## Conclusion -
# Summer Season	and Mansoon Season --> Longer trip Duration
# Autumn Season	and Winter Season --> Shorter trip Duration
```

e. Using multivariate exploration, explain whether there is any relation between short
trips and the users.

```python
short = newdf[newdf['duration_sec'] <= 3600][['duration_sec','user_type']] # 1 Hr as short distance 
long = newdf[newdf['duration_sec'] > 3600][['duration_sec','user_type']]
```

```python
short
```

```python
short['user_type'].value_counts().plot(kind='bar')
```

```python
long['user_type'].value_counts().plot(kind='bar')
```

```python
pd.DataFrame({'short':short['user_type'].value_counts(),'Long':long['user_type'].value_counts()})
```

```python
### During Longer Duration No of Subscribers is more than Customer
### During Shorter Duration No of Customer is more than Subscriber
```

a. Create a figure object and axis of the figure canvas at axis1 = [0,0,1,1]. Provide

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax2.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```
