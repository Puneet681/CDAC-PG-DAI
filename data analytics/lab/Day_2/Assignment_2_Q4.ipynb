```python
UPA = ['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD','KC','RLD','RSP','CMP','KC','PPI','MD','BJP','SS', 'LJP', 'SAD', 'RLSP']
NDA = ['AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK','IJK',
       'PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS','AGP','JSS',
       'PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']

Others = ['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM',
'KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL',
'JVM','MDMK']
```

```python
def Alliance(x):
    if x in UPA:
        x = 'UPA'
    elif x in NDA:
        x = 'NDA'
    else:
        x = 'Others'
    return x
    
can_09['Alliance'] = can_09['Party Abbreviation'].apply(Alliance)
can_09.head()
```

```python
can_14['Alliance'] = can_14['Party Abbreviation'].apply(Alliance)
can_14.head()
```
