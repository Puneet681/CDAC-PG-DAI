```python
a[0] = a[0].fillna(a[0].mean())
a[0] = a[0].round()
a[1] = a[1].fillna(a[1].mean())
a[1] = a[1].round()
a[2] = a[2].fillna(a[3].mean())
a[2] = a[2].round()
a[3] = a[3].fillna(a[3].mean())
a[3] = a[3].round()
a
```

```python
total = df['Gender'].count()
cond = df['Gender'] == 'Male'
Male_count = df[cond]['Gender'].count()
M_per = (Male_count/total)*100
cond = df['Gender'] == 'Female'
Female_count = df[cond]['Gender'].count()
F_per = (Female_count/total)*100

O_count = total-(Male_count+Female_count)
O_per =  100 - (M_per+F_per)

print(f"male count = {Male_count}, male percetage = {M_per}")
print(f"Female count = {Female_count}, Female percentage ={F_per}")
print(f"Other Count = {O_count}, Other percentage = {O_per}")
```

```python
total = df['Gender'].count()
count = df.Gender.value_counts()
ind= count.index

for i in range(len(count)):
    print(f"{ind[i]} count : {count[i]} , {ind[i]} Percentage : {(count[i]/total)*100}")
```

```python
df2 = df[['Age','Price']]
df3 = df2.sort_values('Age')
df4 = df3.groupby('Age').mean()
df4 = df4.reset_index()
```

```python
dict1 = {'Age_range':[],'Avrage':[]}
for i in range(0,len(df4),4):
    key = f"{df4['Age'].iloc[i]} <Age< {df4['Age'].iloc[i]+4}"
    val = df4['Price'].iloc[i:i+4].mean()
    dict1["Age_range"].append(key)
    dict1['Avrage'].append(val)

data = pd.DataFrame(dict1)
data
```

```python
df2 = df[['Age','Price']]
df3 = df2.sort_values('Age')
df4 = df3.groupby('Age').count()
df4 = df4.reset_index()

dict1 = {'Age_range':[],'Avrage':[]}
for i in range(0,len(df4),4):
    key = f"{df4['Age'].iloc[i]} <Age< {df4['Age'].iloc[i]+4}"
    val = df4['Price'].iloc[i:i+4].sum()
    dict1["Age_range"].append(key)
    dict1['Avrage'].append(val)

data = pd.DataFrame(dict1)
data
```
