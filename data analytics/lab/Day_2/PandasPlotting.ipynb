14. Write a Pandas program to create a plot of adjusted closing prices, thirty
days and forty days simple moving average of Alphabet Inc. between two specific
dates.

```python
cond = (df.index>= '2020-4-1') & (df.index<= '2020-9-30')
df1 = df.loc[cond]
df1 = model.fit_transform(df1)
df1 = pd.DataFrame(df1)
plt.figure(figsize=(20,7))
plt.plot(df1[4])
df40 = df1[4].rolling(40).mean()
df30 = df1[4].rolling(30).mean()
sns.lineplot(df40)
sns.lineplot(df30)
```

15. Write a Pandas program to create a plot of adjusted closing prices, 30 days
simple moving average and exponential moving average of Alphabet Inc.
between two specific dates.

```python
cond = (df.index>= '2020-4-1') & (df.index<= '2020-9-30')
df1 = df.loc[cond]
df1 = model.fit_transform(df1)
df1 = pd.DataFrame(df1)
plt.figure(figsize=(20,7))
plt.plot(df1[4])
df30 = df1[4].rolling(30).mean()
dfexm = df1[4].ewm(0.8).mean()
sns.lineplot(df30)
sns.lineplot(dfexm)
```

17. Write a Pandas program to create a plot to visualize daily percentage returns
of Alphabet Inc. stock price between two specific dates.

```python
cond = (df.index>= '2020-4-1') & (df.index<= '2020-9-30')
df1 = df.loc[cond]
df1['Percentage'] = (df1['Adj Close']-df1['Open'])/(df1['Open'])*100
df1
```

```python
plt.figure(figsize=(25,6))
df1['Percentage'].plot(kind='bar')
```

18. Write a Pandas program to plot the volatility over a period of time of Alphabet
Inc. stock price between two specific dates.

```python
cond = (df.index>= '2020-4-1') & (df.index<= '2020-9-30')
df1 = df.loc[cond]
df1 = pd.DataFrame(df1)
df1['Votality'] = df1['Adj Close'].rolling(2).std()
sns.lineplot(df1['Adj Close'])
sns.lineplot(df1['Votality'])
```
