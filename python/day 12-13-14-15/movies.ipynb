### find all masala movie - (action ,romance,comedy,thriller)


```python
target_genres = ['Action', 'Romance', 'Comedy', 'Thriller']
filtered_df = new_mov[new_mov['genres'].apply(lambda x: all(genre in x for genre in target_genres))]
filtered_df
```

```python
#a = [new_mov['genres'].split('|')]
cat = []
for i in new_mov['genres']:
    a = i.split('|')
    cat.append(a)
cat
```

```python
mov = []
for i in new_mov['title']:
    mov.append(i)
mov
```

```python
d = {}
for i in range(len(mov)):
    d[mov[i]] = cat[i]

d
```

```python
# action ,romance,comedy,thriller
dict2 = {}
for k,v in d.items():
    if 'Action' in v and 'Romance' in v and 'Comedy' in v and 'Thriller' in v:
        dict2[k] = "|".join(v)
dict2     
```

```python
# action ,romance,comedy,thriller
new_mov2 = {}
for k,v in d.items():
    if 'Action' in v and 'Romance' in v and 'Comedy' in v and 'Thriller' in v:
        new_mov2[k] = v
        
new_mov2
```
