```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
weather = pd.read_csv('weather_by_cities.csv')
weather
```

```python
cities = weather['city'].unique()
```

```python
grp = weather.groupby('city')
max_temp = []
for i in cities:
    max_temp.append(grp.get_group(i)['temperature'].max())
```

```python
sns.barplot(x=cities,y=max_temp);
```

```python
plt.pie(max_temp,labels=cities);
```

* add headings to the column- sr.no, age, Gender, profession, Views
* display only column gender
* add col6 concatenate values of age and gender and seperate them by :
* retrieve values of age and views display bar graph(hint : agewise grouping and find averageviews)
* retrieve values of profession and views display bar graph (use grouping)

```python
import pandas as pd 
```

```python
movie_data = pd.read_table("http://bit.ly/movieusers",sep="|",header=None)
movie_data
```

```python
movie_data.columns = ['sr.no', 'age','Gender', 'profession', 'Views']
movie_data
```

```python
movie_data['Gender']
```

```python
movie_data['col_6'] = movie_data['age'].astype(str)+movie_data['Gender']
movie_data
```

```python
k = []
for i in movie_data['Views']:
    try:
        k.append(int(i))
    except:
        k.append(0)
        
```

```python
movie_data['Views'] = k
```

```python
movie_data['Views'].astype(int)
```

```python
grp = movie_data.groupby('age')
index = movie_data['age'].unique()
index

avg = []
for i in index:
    avg.append(grp.get_group(i)['Views'].mean())
```

```python
plt.bar(x=index,height=avg)
```

```python
grp = movie_data.groupby('profession')
index = movie_data['profession'].unique()
index

avg = []
for i in index:
    avg.append(grp.get_group(i)['Views'].mean())
```

```python
a = plt.bar(x=index,height=avg,width=0.5)
plt.xticks(rotation=90)
plt.show()
```

```python
a= sns.barplot(x=index,y=avg,)
a.set_xticklabels(a.get_xticklabels(), rotation=90);
```
