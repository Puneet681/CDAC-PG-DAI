```python
# --------- Q3

def add_by_pos(l):
    in_ch = input('''a) add at last position
                        b) add at given position:- 
                        ''')
    if in_ch == 'a':
        val = int(input("enter the value"))
        status = add_at_last(l,val)
        if status == 1:
            print(f"{val} is added in list in {l}")
        else:
            print(f"{val} failed to add in list = {l}")
    if in_ch == 'b':
        val = int(input("enter the value"))
        pos = int(input("enter the position"))
        status = add_at_giv_pos(l,val,pos)
        if status == 1:
            print(f"{val} is added in list in {l}  at {pos}")
        elif status == 0:
            print(f"{val} failed to add in list = {l}  at {pos}")
def add_at_last(l,val):
    l.append(val)
    return 1
def add_at_giv_pos(l,val,pos):
    try:
        l[pos] = val
    except Exception as e:
        print(e)
        return 0
    return 1

def del_by_val(l):
    val = int(input("enter the value"))
    try:
        l.remove(val)
        print(f"{val} deleted from list = {l}")
    except Exception as e:
        print(e,f"\n{val} faild to delete")

def del_pos(l):
    in_ch = input('''a) delete last element
b) delete from particular position
''')
    if in_ch == 'a':
        status = del_at_last(l)
        if status == 1:
            print(f"last element deleted from list == {l}")
        else:
            print(f"fail to remove last element")
    elif in_ch == 'b':
        pos = input("enter position to del")
        status = del_by_pos(l,pos)
        if status == 1:
            print(f"element deleted from list == {l} from {pos}")
        else:
            print(f"fail to remove element from {pos}")
def del_at_last(l):
    try:
        print(l)
        del(l[len(l)-1])
        print(l)
        return 1
    except Exception as e:
        print(e)
        return None
def del_by_pos(l,pos):
    try:
        print([l])
        del(l[pos])
        print([l])
        return 1
    except Exception as e:
        print(e)
        return None
    
def soort(l):
    in_ch = input('''a) ascending
b) descending :- 
''')

    if in_ch == 'a':
        status , l = soort_acc(l)
        if status == 1:
            print(f"list is sorte in accending order = {l}")
        else:
            print(f"fail to sort the list")
    elif in_ch == 'b':
        status ,l = soort_dec(l)
        if status == 1:
            print(f"list is sorted in decending order = {l}")
        else:
            print(f"fail to sort list")
def soort_acc(l):
    l.sorrt()
    return 1,l
def soort_dec(l):
    l.soort(reverse = True)
    return 1,l

def rev(l):
    l.reverse()
    print(f"list has been reversed {l}")

def soort_NL(l):
    g = sorted(l)
    r = [i for i in g]
    print(f"orignal list = {l} \nsorted list = {r}")
    
def rev_NL(l):
    g = reversed(l)
    r = [i for i in g]
    print(f"orignal list = {l} \nreversed list = {r}")

def disp(l):
    in_ch = input('''a) normal
b) numbered:- ''')
    if in_ch == 'a':
        r = disp_nor(l)
        print(f"enterd list = {r}")
    elif in_ch == 'b':
        r = disp_numbered(l)
        print([["index = ",i, "value = ",c] for i,c in enumerate(l)])
    
        print(f" = {r}")
def disp_nor(l):
    r = [i for i in l]
    return r
def disp_numbered(l):
    r = [["index = ",i, "value = ",c] for i,c in enumerate(l)]
    
    return r


def main():
    
    lst = []
    
    try:
        i_str = input("enter the list to do oprations in CSV format")
        i_str = i_str.replace(',',"")
        lst.extend(i_str)
    except Exception as e:
        print(e)
        return None
    
    while True:

        choice = int(input('''enter your choice'
        1. Accept Data
        2. Delete data by value
        3. delete data by position
        4. sort
        5. reverse
        6. Print in sorted order without changing original list
        7. print in reverse order without changing original list
        8. display data choose
        -1. Exit : -'''))

    
            

        if choice == 1 :
            add_by_pos(lst)

        elif choice == 2 :
            del_by_val(lst)

        elif choice == 3 :
            del_pos(lst)
        
        elif choice == 4 :
            soort(lst)
        
        elif choice == 5 :
            rev(lst)
        
        elif choice == 6 :
            soort_NL(lst)
        
        elif choice == 7 :
            rev_NL(lst)
        
        elif choice == 8 :
            disp(lst)

        elif choice == -1:
            print("THANK YOU")
            break

main()
```
